# LSTM

```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, GRU, LSTM, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error


TRAIN_FILE = '/Users/haotianyang/Downloads/archive_1/Google_Stock_Price_Train.csv'
TEST_FILE = '/Users/haotianyang/Downloads/archive_1/Google_Stock_Price_Test.csv'
FEATURES = ['Open', 'High', 'Low', 'Close', 'Volume']

SEQUENCE_LENGTH = 30   
PREDICT_HORIZON = 1    
BATCH_SIZE = 32
EPOCHS = 50
VALIDATION_SPLIT = 0.15
LEARNING_RATE = 0.01 # Change Learning rate here!!!
MODEL_TYPE = 'LSTM' # Switch to different model here!!!

CLOSE_INDEX = FEATURES.index('Close')  

train_df = pd.read_csv(TRAIN_FILE)
test_df = pd.read_csv(TEST_FILE)

print("Missing values in train:\n", train_df.isna().sum())
print("Missing values in test:\n", test_df.isna().sum())

for col in FEATURES:
    train_df[col] = train_df[col].astype(str).str.replace(',', '')
    test_df[col] = test_df[col].astype(str).str.replace(',', '')
    train_df[col] = train_df[col].astype(float)
    test_df[col] = test_df[col].astype(float)

train_data = train_df[FEATURES].values
test_data = test_df[FEATURES].values

# Data scale
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train_data)

# Sequences
def create_sequences(data, seq_length=30, pred_horizon=1):
    X, y = [], []
    for i in range(len(data) - seq_length - pred_horizon + 1):
        X.append(data[i:(i+seq_length)])  # shape: (seq_length, features)
        # y: next day's Close price
        y.append(data[i+seq_length : i+seq_length+pred_horizon, CLOSE_INDEX])
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_scaled, SEQUENCE_LENGTH, PREDICT_HORIZON)
y_train = y_train.reshape(-1)  # shape (samples,)

print("X_train shape:", X_train.shape)  
print("y_train shape:", y_train.shape)

# Model Function
def build_model(model_type='LSTM', learning_rate=0.001):
    model = Sequential()
    if model_type == 'SimpleRNN':
        model.add(SimpleRNN(64, return_sequences=True, input_shape=(SEQUENCE_LENGTH, len(FEATURES))))
        model.add(SimpleRNN(64))
    elif model_type == 'GRU':
        model.add(GRU(64, return_sequences=True, input_shape=(SEQUENCE_LENGTH, len(FEATURES))))
        model.add(GRU(64))
    else:
        # Default to LSTM
        model.add(LSTM(64, return_sequences=True, input_shape=(SEQUENCE_LENGTH, len(FEATURES))))
        model.add(LSTM(64))
    
    model.add(Dense(1))  # Only predict Close price
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

model = build_model(MODEL_TYPE, LEARNING_RATE)
model.summary()

# Training 
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)

history = model.fit(
    X_train, y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VALIDATION_SPLIT,
    shuffle=True,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title(f'{MODEL_TYPE} Model Loss Over Epochs (Predicting Close)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()


dataset_total = np.concatenate((train_data, test_data), axis=0)
inputs = dataset_total[len(dataset_total) - len(test_data) - SEQUENCE_LENGTH:]
inputs_scaled = scaler.transform(inputs)

X_test = []
for i in range(SEQUENCE_LENGTH, SEQUENCE_LENGTH + len(test_data)):
    X_test.append(inputs_scaled[i-SEQUENCE_LENGTH:i])
X_test = np.array(X_test)

print("X_test shape:", X_test.shape)

actual_close = test_data[PREDICT_HORIZON-1:, CLOSE_INDEX]

# Prediction
predicted_scaled = model.predict(X_test)

# Create a dummy array for inverse transform
temp = np.zeros((predicted_scaled.shape[0], len(FEATURES)))
temp[:, CLOSE_INDEX] = predicted_scaled[:, 0]

predicted_all_features = scaler.inverse_transform(temp)
predicted_close = predicted_all_features[:, CLOSE_INDEX]

# ---------------------------
# Step 8: Evaluation (MSE, RMSE, MAPE)
# ---------------------------
mse = mean_squared_error(actual_close, predicted_close)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(actual_close, predicted_close)

print(f"{MODEL_TYPE} Test MSE (Close Only): {mse}")
print(f"{MODEL_TYPE} Test RMSE (Close Only): {rmse}")
print(f"{MODEL_TYPE} Test MAPE (Close Only): {mape * 100}%")

# Plot comparison for the close price
plt.figure()
plt.plot(actual_close, color='red', label='Real Google Close Price')
plt.plot(predicted_close, color='blue', label='Predicted Google Close Price')
plt.title(f'Google Close Price Prediction with {MODEL_TYPE}')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
